## Dataset Overview and Executive Summary

the dataset is Ford gobike service which is a dataset for the serivce that Ford lunched in San Francisco.
in this analysis I found that one time customer is way more than subscribers and the duration for each gender where women are more than men when using the service and the type of users who share their bikes and it wasn't the subscribers but the customers

```python
# import all packages and set plots to be embedded inline
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# suppress warnings from final output
import warnings
warnings.simplefilter("ignore")
```

```python
# load in the dataset into a pandas dataframe
df = pd.read_csv('cleaned_dataset_for_ford_gobike.csv');
```

## (Visualization 1)

here in this plot will be showing the ratio between subscribers and the customers and the customers are way more than subscribers 

```python
x = df['user_type'].unique()
y = df['user_type'].value_counts(normalize=True)


plt.bar(x,y);
plt.title('Bar chart of user type using the service');
```

## (Visualization 2)

this box plot compare the average between the genders and we can see that females having more duration more than males

```python
sns.boxplot(data=df,x='member_gender',y='duration_sec',color='tab:blue');
plt.xlabel('Gender')
plt.ylabel('Duration')
plt.title('Box plot about the gender and thire duration');
```

## (Visualization 3)

this clustered bar chart is comparing  which users type is sharing bike while comparing to their density as we can see the only users type that sharing the bike is the customer

```python
sns.countplot(data=df,x='bike_share_for_all_trip',hue='user_type');
plt.xlabel('Bike Share')
plt.ylabel('Density')
plt.title('Culstered Bar Plot of  the enrollment type and type of users ');
```
